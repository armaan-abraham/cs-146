```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

def get_likelihood(vars, theta):
  acc = 1
  for var in vars:
    acc *= theta ** var * (1 - theta) ** (1-var)
  return acc

PLOT_I = 1

def plot(n, n_1):
  global PLOT_I
  step = 0.01
  thetas = list(np.arange(0, 1 + step, step))
  vars = [1] * n_1 + [0] * (n - n_1)
  likelihoods = [get_likelihood(vars, theta) for theta in thetas]
  print(sum(likelihoods))
  ax = plt.gca()
  ax.set_xlim([0, 1])
  plt.figure(PLOT_I)
  plt.scatter(thetas, likelihoods)
  plt.title(f"L(θ) for {n_1} 1s and {n-n_1} 0s")
  plt.ylabel("L(θ)")
  plt.xlabel("θ")
  PLOT_I += 1
  print(f"Max likelihood θ: {thetas[likelihoods.index(max(likelihoods))]}")

plot(10, 6)
plot(5, 3)
plot(100, 60)
plot(10,5)
```
